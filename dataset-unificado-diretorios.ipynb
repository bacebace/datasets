# Bubliotecas

```python
from pathlib import Path
import os
import pandas as pd

import pickle

from sklearn.model_selection import train_test_split

import shutil
```

# Criar dataframe a partir do dataset

```python
dataset="/kaggle/input/garbage-classification-u/garbage_classification_u"
#dataset="/kaggle/input/garbage-classification/garbage_classification"
image_dir = Path(dataset)

# Get filepaths and labels
filepaths = list(image_dir.glob(r'**/*.JPG')) + list(image_dir.glob(r'**/*.jpg')) + list(image_dir.glob(r'**/*.png')) + list(image_dir.glob(r'**/*.png'))

labels = list(map(lambda x: os.path.split(os.path.split(x)[0])[1], filepaths))

filepaths = pd.Series(filepaths, name='Filepath').astype(str)
labels = pd.Series(labels, name='Label')

# Concatenate filepaths and labels
image_df = pd.concat([filepaths, labels], axis=1)
```

```python
image_df
```

# Separar em conjuntos de treino e teste

```python
train_df1, test_df = train_test_split(image_df, test_size=0.2, shuffle=True, random_state=42)
```

```python
train_df1.shape
```

```python
test_df.shape
```

# Criar o conjunto de validação a partir do conjunto de treino

```python
train_df, val_df = train_test_split(train_df1, test_size=0.2, shuffle=True, random_state=42)
```

```python
train_df.shape
```

```python
val_df.shape
```

# Criar diretórios para cada conjunto

```python
os.mkdir("/kaggle/working/train")
os.mkdir("/kaggle/working/val")
os.mkdir("/kaggle/working/test")
```

# Copiar as imagens para seu respectivo diretório

```python
train_path = Path("/kaggle/working/train")
val_path = Path("/kaggle/working/val")
test_path = Path("/kaggle/working/test")
```

```python
for image in train_df.Filepath:
    shutil.copy(image, train_path)
    
for image in val_df.Filepath:
    shutil.copy(image, val_path)
    
for image in test_df.Filepath:
    shutil.copy(image, test_path)
```
